```python
lines_points = []
for line in lines:
  cities = line.split("-")
  for city_code in cities: 
    city_coords = (float(cities_airports[cities_airports["IATA"] == city_code]["lat"]),float(cities_airports[cities_airports["IATA"] == city_code]["lng"]))
    lines_points.append(city_coords)
```

```python
lines_points
```

```python
# import folium library
import folium

# Creates map object
map = folium.Map(location=[4,-74], tiles="OpenStreetMap", zoom_start=5)

for city in range(0, n_cities):
       folium.Marker(location=[cities_airports.iloc[city]['lat'], cities_airports.iloc[city]['lng']],popup = "-Ciudad : " + cities_airports.iloc[city]['city'] + "\n" + " -Departamento : " + cities_airports.iloc[city]['admin_name']  + "\n" + "-Codigo ciudad : " + cities_airports.iloc[city]['IATA']).add_to(map)

lines = folium.PolyLine(lines_points).add_to(map)

# Show the map
map
```
